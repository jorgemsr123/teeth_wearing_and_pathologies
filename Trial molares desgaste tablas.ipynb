```python
# Seleccion de solo molares


max_der_num=range(17,15,-1)
# ubicacion dientes maxilar derecho
# I need to give the first column to have a template, after it the join works 

max_izq_num=range(26,29,1)
#ubicacion dientes maxilar izquierdo

mand_der_num=range(48,45,-1)
## ubicacion dientes mandibular derecho

mand_izq_num=range(36,39,1)
## ubicacion dientes mandibular izquierdo

teeth_num=[max_der_num,max_izq_num, mand_der_num,mand_izq_num]
# four elements each a range given the location of some teeth
```

```python
up_down=['Maxilar','Mandíbula']

side_maxilar=['Derecho','Izquierdo']

side_mandibular=['Derecho ','Izquierdo']
```

```python
#this is the way to make a table just with one of the characteristics of the teeth


df_g= df_calculo.xs(('Maxilar','Derecho',18), axis=1, level=[0,1,2])

for place in up_down:
    if place=='Maxilar':
        for side in side_maxilar:
            if side =='Derecho':
                for num in max_der_num:
                    df_g=df_g.join(df_calculo.xs((place,side,num), axis=1, level=[0,1,2]))
            elif side == 'Izquierdo':
                for num in max_izq_num:
                    df_g=df_g.join(df_calculo.xs((place,side,num), axis=1, level=[0,1,2]))
    elif place=='Mandíbula':
        for side in side_mandibular:
            if side =='Derecho ':
                for num in mand_der_num:
                    df_g=df_g.join(df_calculo.xs((place,side,num), axis=1, level=[0,1,2]))
            elif side == 'Izquierdo':
                for num in mand_izq_num:
                    df_g=df_g.join(df_calculo.xs((place,side,num), axis=1, level=[0,1,2]))
```
