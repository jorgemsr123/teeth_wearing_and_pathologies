```python
# Hue lugar profundidad caries

# First I need to define car_P where I have the melt of the first column I need, 
# after I have to initialize the extra column where I am going to locate Rango Edad
# At that moment I can start the for to complete Rango Edad for the first columns

car_P=pd.melt(pd.DataFrame(df.iloc[:,53]))
car_P['Lugar']=0
q=-1
for item in df['Lugar']:
    q=q+1
    car_P['Lugar'][q]=df['Lugar'][q]
        

# When the for is finish the first columns is ready
# Now it is possible to add the other columns I need. 
# Without the first columns ready, it is not possible to add others


# I started a for that include the columns I need without including the one that is ready
# I need to do the same as in the first column but with two for
# The first FOR is to select the column and start with the melt
# The second FOR is to include the lable Rango edad


for a in range(54,57,1):
    car=(pd.melt(pd.DataFrame(df.iloc[:,a])))
    car['Lugar']=a
    i=-1
    for item in df['Lugar']:
        i=i+1
        car['Lugar'][i]=df['Lugar'][i]
    car_P=car_P.append(car)
```

```python
# It is nicer if the zeros in values are not present 

nz= car_P['value']!=0

car_PT=car_P[nz]

car_PT
```

```python
palette_lugar={'Palestina':'blue','Chinchiná':'orange'}
```

```python
plt.figure(figsize=(6,4))

sns.swarmplot(x='variable', y='value', hue='Lugar', data=car_PT,hue_order=['Palestina','Chinchiná'],palette=palette_lugar)


plt.xlabel('Diferentes profundidades de las caries')
plt.ylabel('Número de caries')
plt.title('Distribución de la profundidad de las caries por lugar')

plt.savefig('//home/jorgemsr123/Desktop/Pavita/CSV_tables/Figures/final/ hue_lugar_caries_por_profundidad.png', dpi=300, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format='png',
        transparent=False, bbox_inches=None, pad_inches=0.1, metadata=None)
```
