```python
# Hue Rango Edad profundidad caries

# First I need to define car_P where I have the melt of the first column I need, 
# after I have to initialize the extra column where I am going to locate Rango Edad
# At that moment I can start the for to complete Rango Edad for the first columns

car_P=pd.melt(pd.DataFrame(df.iloc[:,52]))
car_P['Rango Edad']=0
q=-1
for item in df['Rango Edad']:
        q=q+1
        car_P['Rango Edad'][q]=df['Rango Edad'][q]
        
# When the for is finish the first columns is ready
# Now it is possible to add the other columns I need. 
# Without the first columns ready, it is not possible to add others


# I started a for that include the columns I need without including the one that is ready
# I need to do the same as in the first column but with two for
# The first FOR is to select the column and start with the melt
# The second FOR is to include the lable Rango edad


for a in range(53,56,1):
    car=(pd.melt(pd.DataFrame(df.iloc[:,a])))
    car['Rango Edad']=a
    i=-1
    for item in df['Rango Edad']:
        i=i+1
        car['Rango Edad'][i]=df['Rango Edad'][i]
    car_P=car_P.append(car)
        
```
